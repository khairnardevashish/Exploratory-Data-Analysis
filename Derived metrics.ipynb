## Analysing ODI cricket data is always exciting. The following data set contains the performance information of each player who played between 1971 - 2011. 

```python
df = pd.read_csv("odi-batting.csv")
```

```python
df.head(10)
```

```python
df.shape
```

##### Which player scored the highest number of centuries?

```python
df["Century"] = df["Runs"].apply(lambda x: 1 if x >= 100 else 0)
df["Year"] = df["MatchDate"].str.split('-', expand= True)[2]
```

```python
table = pd.pivot_table(df, values='Century', index=['Player','Country'], aggfunc = np.sum)
```

```python
table.sort_values(by=['Century'],ascending=False)
```

### Scenario 1
The ‘strike rate’ of a batsman is defined as the (number of runs scored / number of 100 balls faced). It is a measure of ‘how fast a batsman has played’.For example, if a batsman scores 40 runs in 50 balls, the strike rate is 80 runs/100 balls.

Among all the innings where a century was scored by a batsman, which one has the highest strike rate?

```python
df["SR"] = round((df["Runs"]/df["Balls"])*100,2)
```

```python
df2 = df[df.Century >= 1]
```

```python
table = pd.pivot_table(data = df2, index = ['Player', 'SR'], values = 'Century', aggfunc = np.sum)
```

```python
table.sort_values(by = ['SR'], ascending = False)
```
