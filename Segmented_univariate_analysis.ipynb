##### What percentage of females in the age group 20-24 are illiterate in India, i.e. out of all the females in the age group 20-24, what fraction is illiterate?

```python
df["Age-group"].value_counts()
```

```python
df.groupby(df["Age-group"])["Illiterate Females"].sum()
```

From above we have number of illiterate women in age 20-24 is 48849864

```python
df.groupby(df["Age-group"])["Total Females"].sum()
```

Total Women in 20-24 is 215358116

Hence percentage is


```python
print("percentage of females in the age group 20-24 are illiterate in India is ", round(48849864/215358116,2)*100,"%")
```

##### Compare the literacy rates (i.e. the number of literates / total number of population) in each age group and choose the correct option.

```python
# Added new column for literacy rate and plot boxplot across it
df["Literacy rate"] = df["Literate Persons"] / df["Total Persons"] *100
```

```python
df.head()
```

```python
sns.catplot(x=df["Age-group"], y = df["Literacy rate"], data =df, kind="box", height =20)
```
